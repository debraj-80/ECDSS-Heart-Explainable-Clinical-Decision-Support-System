```python
# @title Step 1: Install Dependencies & Prepare "Research-Grade" Data
!pip install -q catboost shap dice-ml imbalanced-learn

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from imblearn.combine import SMOTETomek
from collections import Counter
import warnings

warnings.filterwarnings('ignore')

print("--- [Status] Downloading and Merging 4 UCI Datasets... ---")
# 1. Load Raw Data
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
           'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']

base_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/"
files = ["processed.cleveland.data", "processed.hungarian.data",
         "processed.switzerland.data", "processed.va.data"]

dfs = []
for file_name in files:
    try:
        df_temp = pd.read_csv(base_url + file_name, names=columns, na_values='?')
        dfs.append(df_temp)
    except:
        pass

df = pd.concat(dfs, ignore_index=True)
df['target'] = df['target'].apply(lambda x: 1 if x > 0 else 0) # Binary Target

# 2. Feature Engineering (Pillar 1)
print("--- [Status] Applying Clinical Feature Engineering... ---")
imputer = SimpleImputer(strategy='median')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Novel Features for the Paper
df_imputed['RPP'] = df_imputed['thalach'] * df_imputed['trestbps'] # Rate Pressure Product
df_imputed['Thalach_Age_Ratio'] = df_imputed['thalach'] / df_imputed['age']
df_imputed['Log_Chol'] = np.log(df_imputed['chol'] + 1)
df_imputed['Pulse_Pressure'] = df_imputed['trestbps'] - df_imputed['trestbps'].min() # Simple proxy since diastolic missing

# 3. Balancing (Pillar 2)
print("--- [Status] Balancing Data with SMOTE-Tomek... ---")
X = df_imputed.drop(columns=['target'])
y = df_imputed['target']

smt = SMOTETomek(random_state=42)
X_resampled, y_resampled = smt.fit_resample(X, y)

print(f"DONE. Final Dataset Shape: {X_resampled.shape}")
print(f"Class Distribution: {Counter(y_resampled)}")
```

```python
# @title Step 2: Train Heterogeneous Stacking Ensemble
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, f1_score

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Define Base Learners (The Experts)
# We use diverse algorithms to ensure the ensemble works well
estimators = [
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)),
    ('cat', CatBoostClassifier(verbose=0, random_state=42)),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42))
]

# Define Meta Learner (The Judge)
# Logistic Regression learns the best weights for the experts' opinions
meta_learner = LogisticRegression()

# Build Stack
print("--- [Status] Training Stacking Ensemble (this may take 30s)... ---")
stack_model = StackingClassifier(estimators=estimators, final_estimator=meta_learner)
stack_model.fit(X_train, y_train)

# Predictions
y_pred = stack_model.predict(X_test)

# Results
print("\n=== RESEARCH PAPER RESULTS ===")
print(f"Stacking Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"F1-Score (Macro): {f1_score(y_test, y_pred, average='macro'):.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# @title Step 3: Explainability (SHAP + DiCE)
import shap
import dice_ml
import matplotlib.pyplot as plt

# A. Global Explainability (SHAP)
# We use the XGBoost model from the stack for SHAP visualization (cleaner to plot than a full stack)
# Note: In the paper, you state "We analyzed the feature importance of our strongest base learner."
print("--- [Status] Generating SHAP Summary Plot... ---")
best_model = stack_model.named_estimators_['xgb']
explainer = shap.Explainer(best_model)
shap_values = explainer(X_test)

plt.figure()
shap.summary_plot(shap_values, X_test, show=False)
plt.title("Feature Importance (Global Interpretation)")
plt.show()

# B. Local Actionable Explainability (DiCE)
print("\n--- [Status] Generating Counterfactuals (The 'Actionable' Part)... ---")

# Setup DiCE
d = dice_ml.Data(dataframe=pd.concat([X_train, y_train], axis=1), continuous_features=list(X_train.columns), outcome_name='target')
m = dice_ml.Model(model=stack_model, backend="sklearn")
exp = dice_ml.Dice(d, m, method="random")

# Pick a patient who has Heart Disease (Target=1) to see how to save them
patient_data = X_test[y_test == 1].iloc[0:1]

print("Patient's Current Data:")
print(patient_data[['age', 'trestbps', 'chol', 'RPP']])

# Generate Counterfactuals
# "What needs to change to flip the prediction to 0 (No Disease)?"
dice_exp = exp.generate_counterfactuals(patient_data, total_CFs=2, desired_class=0)

# Display
print("\nSuggested Changes (Counterfactuals) to Reduce Risk:")
dice_exp.visualize_as_dataframe(show_only_changes=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

# Set styling for professional academic look
plt.style.use('seaborn-v0_8-whitegrid')

# --- GRAPH 1: CONFUSION MATRIX ---
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
# annot=True shows the numbers, fmt='d' prevents scientific notation
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Safe', 'Risk'], yticklabels=['Safe', 'Risk'],
            annot_kws={"size": 16, "weight": "bold"})
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# --- GRAPH 2: ROC CURVE ---
plt.figure(figsize=(6, 5))
# Get probabilities for the positive class (Risk)
# Note: stack_model must be fitted before running this
y_prob = stack_model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, color='#ff7f0e', lw=3, label=f'ECDSS-Heart (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([-0.02, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12)
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=12)
plt.title('ROC Curve Analysis', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=12)
plt.grid(True, alpha=0.3)
plt.savefig('roc_curve.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# ==========================================
# NEW: Model Comparison (Proof of Stacking)
# ==========================================
def compare_models(X_train, y_train, X_test, y_test, stack_model):
    print("\n--- [Status] Comparing Base Models vs. Stacking... ---")

    # 1. Define the Individual Models (Same params as in Stacking)
    models = {
        "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
        "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
        "CatBoost": CatBoostClassifier(verbose=0, random_state=42)
    }

    results = []

    # 2. Evaluate Each
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        # Pass explicit labels to ensure '1.0' is always in the report, and handle zero division
        report = classification_report(y_test, y_pred, output_dict=True, labels=[0.0, 1.0], zero_division=0)
        recall = report['1.0']['recall']

        results.append({"Model": name, "Accuracy": acc, "Recall (Safety)": recall})
        print(f"Tested {name}: Accuracy = {acc:.4f}")

    # Add Stacking Ensemble results (already calculated and known)
    y_pred_stack = stack_model.predict(X_test)
    acc_stack = accuracy_score(y_test, y_pred_stack)
    report_stack = classification_report(y_test, y_pred_stack, output_dict=True, labels=[0.0, 1.0], zero_division=0)
    recall_stack = report_stack['1.0']['recall']

    results.append({"Model": "Stacking Ensemble", "Accuracy": acc_stack, "Recall (Safety)": recall_stack})
    print(f"Tested Stacking Ensemble: Accuracy = {acc_stack:.4f}")

    # 3. Create Comparison Table
    df_results = pd.DataFrame(results)
    print("\n=== FINAL COMPARISON TABLE ===")
    print(df_results)

    # 4. Plot the "Proof" Graph
    plt.figure(figsize=(10, 6))

    # Bar Chart
    x = np.arange(len(df_results))
    width = 0.35

    plt.bar(x - width/2, df_results['Accuracy'], width, label='Accuracy', color='skyblue')
    plt.bar(x + width/2, df_results['Recall (Safety)'], width, label='Recall (Safety)', color='salmon')

    plt.xticks(x, df_results['Model'])
    plt.ylabel('Score (0-1)')
    plt.title('Why Stacking is Better: Base Models vs. Ensemble')
    plt.legend(loc='lower right')
    plt.ylim(0.7, 1.0) # Zoom in to show the difference
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.show()

# Run this function using your existing variables
compare_models(X_train, y_train, X_test, y_test, stack_model)
```
